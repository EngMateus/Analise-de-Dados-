#### QUAL É O PREÇO MÉDIO DE ALUGUEL POR TIPO DE IMÓVEL E BAIRRO? PODEMOS IDENTIFICAR OS BAIRROS MAIS CAROS E OS MAIS ACESSÍVEIS?

```python
precos = imoveis_residenciais.groupby(['Bairro', 'Tipo'])['Valor'].mean().reset_index()
```

```python
bairro_media = precos.groupby('Bairro')['Valor'].mean().reset_index()
```

```python
top_5_caros = bairro_media.nlargest(5, 'Valor')
top_5_baratos = bairro_media.nsmallest(5, 'Valor')
```

#### QUANTIDADE DE IMOVEIS TOP 10 BAIRRO MAIS MOVIMENTADOS

```python
imoveis_por_bairro = imoveis_residenciais.groupby(['Bairro', 'Tipo']).size().reset_index(name='Quantidade')
### Pegando os 10 bairros com maior quantidade de imoveis
top_bairros_selecionados = imoveis_por_bairro.groupby('Bairro')['Quantidade'].sum().nlargest(10).index
top_imoveis_por_bairro = imoveis_por_bairro[imoveis_por_bairro['Bairro'].isin(top_bairros_selecionados)]
```

```python
plt.figure(figsize=(16, 8))
sns.barplot(data=top_imoveis_por_bairro, x='Bairro', y='Quantidade', hue='Tipo', palette='Set3')
plt.title('Quantidade de Cada Tipo de Imóvel nos Top 10 Bairros')
plt.xlabel('Bairro')
plt.ylabel('Quantidade de Imóveis')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tipo de Imóvel', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### BAIRROS MAIS CAROS E MAIS BARATOS POR TIPO DE IMOVEL

```python
top_bairros = precos[precos['Bairro'].isin(top_5_caros['Bairro']) | precos['Bairro'].isin(top_5_baratos['Bairro'])]
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=top_bairros, x='Bairro', y='Valor', hue='Tipo', palette='viridis')
plt.title('Top 5 Bairros Mais Caros e Mais Baratos por Tipo de Imóvel')
plt.xlabel('Bairro')
plt.ylabel('Preço Médio de Aluguel')
plt.xticks(rotation=45)
plt.legend(title='Tipo de Imóvel')
plt.show()
```

#### EXISTE UMA CORRELAÇÃO SIGNIFICATIVA ENTRE A ÁREA DO IMÓVEL E O VALOR DO ALUGUEL?

```python
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=imoveis_residenciais, x='Area', y='Valor', alpha=0.7)
sns.regplot(data=imoveis_residenciais, x='Area', y='Valor', scatter=False, color='r', line_kws={'label': 'Linha de Tendência'})
plt.title('Correlação Entre Área do Imóvel e Valor do Aluguel')
plt.xlabel('Área do Imóvel (m²)')
plt.ylabel('Valor do Aluguel')
plt.show()
```

#### TOP 10 BAIRROS MAIS CAROS

```python
media_aluguel_por_bairro = imoveis_residenciais.groupby('Bairro')['Valor'].mean().reset_index()
```

```python
top_10_bairros = media_aluguel_por_bairro.nlargest(10, 'Valor')['Bairro']
```

```python
top_bairros_df = imoveis_residenciais[imoveis_residenciais['Bairro'].isin(top_10_bairros)]
```

```python
tabela_pivot = top_bairros_df.pivot_table(values='Valor', index='Bairro', columns='Tipo', aggfunc='mean')
```

```python
plt.figure(figsize=(14, 8))
sns.heatmap(tabela_pivot, cmap='coolwarm', annot=True, fmt='.2f', linewidths=.5)
plt.title('Heatmap dos Top 10 Bairros Mais Caros por Tipo de Imóvel')
plt.xlabel('Tipo de Imóvel')
plt.ylabel('Bairro')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
